```python
def new_team (row):
    if row['Tm'] == 'Arizona Cardinals':
        return 'ARI'
    if row['Tm'] == 'Atlanta Falcons':
        return 'ATL'
    if row['Tm'] == 'Baltimore Ravens':
        return 'BAL'
    if row['Tm'] == 'Buffalo Bills':
        return 'BUF'
    if row['Tm'] == 'Carolina Panthers':
        return 'CAR'
    if row['Tm'] == 'Chicago Bears':
        return 'CHI'
    if row['Tm'] == 'Cincinnati Bengals':
        return 'CIN'
    if row['Tm'] == 'Cleveland Browns':
        return 'CLE'
    if row['Tm'] == 'Dallas Cowboys':
        return 'DAL'
    if row['Tm'] == 'Denver Broncos':
        return 'DEN'
    if row['Tm'] == 'Detroit Lions':
        return 'DET'
    if row['Tm'] == 'Green Bay Packers':
        return 'GB'
    if row['Tm'] == 'Houston Texans':
        return 'HOU'
    if row['Tm'] == 'Indianapolis Colts':
        return 'IND'
    if row['Tm'] == 'Jacksonville Jaguars':
        return 'JAC'
    if row['Tm'] == 'Kansas City Chiefs':
        return 'KC'
    if row['Tm'] == 'Los Angeles Chargers':
        return 'LAC'
    if row['Tm'] == 'Los Angeles Rams':
        return 'LA'
    if row['Tm'] == 'Miami Dolphins':
        return 'MIA'
    if row['Tm'] == 'Minnesota Vikings':
        return 'MIN'
    if row['Tm'] == 'New England Patriots':
        return 'NE'
    if row['Tm'] == 'New Orleans Saints':
        return 'NO'
    if row['Tm'] == 'New York Giants':
        return 'NYG'
    if row['Tm'] == 'New York Jets':
        return 'NYJ'
    if row['Tm'] == 'Oakland Raiders':
        return 'OAK'
    if row['Tm'] == 'Philadelphia Eagles':
        return 'PHI'
    if row['Tm'] == 'Pittsburgh Steelers':
        return 'PIT'
    if row['Tm'] == 'San Diego Chargers':
        return 'SD'
    if row['Tm'] == 'San Francisco 49ers':
        return 'SF'
    if row['Tm'] == 'Seattle Seahawks':
        return 'SEA'
    if row['Tm'] == 'St. Louis Rams':
        return 'STL'
    if row['Tm'] == 'Tampa Bay Buccaneers':
        return 'TB'
    if row['Tm'] == 'Tennessee Titans':
        return 'TEN'
    if row['Tm'] == 'Washington Redskins':
        return 'WAS'
    return 'Other'
```

```python
winning_pct['Team'] = winning_pct.apply (lambda row: new_team(row), axis=1)
```
